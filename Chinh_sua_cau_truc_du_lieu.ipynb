# 0. Import thư viện pandas

```python
import pandas as pd
```

# 1. Đọc dữ liệu từ file CSV vào Dataframe

```python
df = pd.read_csv("https://s3.ap-southeast-1.amazonaws.com/learn.codegym.vn/media/public/2021/12/31021019/FoodPrice_in_Turkey.csv")
df.head()
```

# 2. Đổi tên cột thuộc tính

```python
# Đổi tên cột "Place" thành "Địa điểm", "ProductName" thành "Tên SP"
df.rename(columns = {"Place": "Địa điểm", "ProductName": "Tên SP"}, inplace = True)
df.head()
```

# 3. Thêm một cột mới trong DataFrame

```python
# Thêm một cột mới với tất cả các giá trị rỗng NaN
```

```python
df["new_column"] = "NaN"
df.head()
```

```python
# Thêm cột giảm giá 10% cho tất cả các bản ghi
# Cách 1: Gán tên cột dưới dạng một chuỗi và thêm giá trị vào cột đó
```

```python
df["Giảm giá"] = pd.Series("10%", index = df.index)
df.head()
```

```python
# Thêm cột giảm giá 12% cho tất cả các bản ghi
# Cách 2: Sử dụng phương thức insert() gồm 3 đối số
# - Đối số đầu tiên chỉ vị trí muốn chèn cột mới, nếu giá trị này là 10 thì cột mới sẽ ngay sau cột thứ 10
# - Đối số thứ hai là tên của cột mới muốn chèn
# - Đối số thứ 3 là giá trị của cột

df.insert(9, "Giảm giá N", pd.Series("12%", index = df.index))
df.head()
```

# 4. Thêm dòng mới vào trong DataFrame

```python
# Sử dụng phương thức append() thêm một dòng vào cuối DataFrame
df = df.append({"Địa điểm": "Nghệ An",
                "ProductId":"RR",
                "Tên SP": "Nút chuối",
                "UmId": "10",
                "UmName": "KG",
                "Month": 6,
                "Year": 2021,
                "Price": 5,
                "Giảm giá": "10%",
                "Giảm giá N": "12%"
               }, ignore_index = True)
df.tail()
```

# 5. Xoá một cột trong DataFrame

```python
# Xoá một cột sử dụng hàm del
```

```python
del df["new_column"]
df.head()
```

```python
# Xoá một cột sử dụng phương thức pop().
# Phương thức pop() trả về cột đã xoá
```

```python
df.pop("Giảm giá N")
```

```python
df.head()
```

```python
# Xoá một cột sử dụng phương thức drop()
# Cú pháp: df.drop("column_name", axis = 1, inplace = True)
# Trong đó axis = 1 là xoá cột; inplace = True xoá trực tiếp trên dữ liệu gốc mà không phải tạo ra bản sao

df.drop("Giảm giá", axis = 1, inplace = True)
df.head()
```

```python
# Xoá nhiều cột sử dụng phương thức drop()

df.drop(["Month", "Year"], axis = 1, inplace = True)
df.head()
```

# 6. Xoá các dòng trong DataFrame

```python
# Cú pháp: df.drop(Chỉ số dòng cần xoá, axis = 0, inplace = True)
# axis = 0 là giá trị mặc định, có thể viết tường minh hoặc không
# Xoá dòng có chỉ số 1 (dòng thứ 2) sử dụng phương thức drop()

df.drop(1, axis = 0, inplace = True)
```

```python
df.head()
```

```python
# Xoá nhiều dòng sử dụng phương thức drop()
# Xoá các dòng có chỉ số 7380 & 7381 như sau:

df.drop([7380, 7381], inplace = True)
df.tail()
```
